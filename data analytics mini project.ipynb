```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample data
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],
    'Income': [0, 3, 4100, 4300, 4400, 4500, 4600, 4700, 4800, 4900, 5000, 5100],
    'Expenses': [2500, 3200, 3100, 3300, 3400, 3500, 3600, 3700, 3800, 3900, 4000, 4100]
}

df = pd.DataFrame(data)
df['Savings'] = df['Income'] - df['Expenses']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame creation for demonstration (if df is not already defined)
data = {
    'Income': [5000, 6000, 7000, 8000, 9000],
    'Expenses': [3000, 3200, 3400, 3600, 3800],
    'Savings': [2000, 2800, 3600, 4400, 5200]
}
df = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df[['Income', 'Expenses', 'Savings']].corr()

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt='.2f')
plt.title('Correlation Heatmap of Financial Metrics')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July'],
    'Income': [5000, 5200, 4800, 5300, 5400, 5500, 5600],
    'Expenses': [4000, 4200, 3900, 4100, 4300, 4400, 4500]
}
df = pd.DataFrame(data)

# Aggregate the total income and expenses
total_income = df['Income'].sum()
total_expenses = df['Expenses'].sum()

# Data for the pie chart
labels = ['Income', 'Expenses']
sizes = [total_income, total_expenses]
colors = ['#4CAF50', '#FF6347']  # Green for Income, Red for Expenses
explode = (0.1, 0)  # Explode the first slice (Income)

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Total Income vs. Expenses')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July'],
    'Income': [5000, 5200, 4800, 5300, 5400, 5500, 5600],
    'Expenses': [4000, 4200, 3900, 4100, 4300, 4400, 4500]
}
df = pd.DataFrame(data)

# Plotting the histograms
plt.figure(figsize=(10, 6))

# Histogram for Income
plt.hist(df['Income'], bins=7, alpha=0.5, label='Income', color='b', edgecolor='black')

# Histogram for Expenses
plt.hist(df['Expenses'], bins=7, alpha=0.5, label='Expenses', color='r', edgecolor='black')

plt.xlabel('Amount ($)')
plt.ylabel('Frequency')
plt.title('Histogram of Monthly Income and Expenses')
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July'],
    'Income': [5000, 5200, 4800, 5300, 5400, 5500, 5600],
    'Expenses': [4000, 4200, 3900, 4100, 4300, 4400, 4500]
}
df = pd.DataFrame(data)

# Preparing data for the box plot
data_to_plot = [df['Income'], df['Expenses']]

# Creating the box plot
plt.figure(figsize=(10, 6))
plt.boxplot(data_to_plot, patch_artist=True, labels=['Income', 'Expenses'])

# Adding titles and labels
plt.title('Box Plot of Monthly Income and Expenses')
plt.ylabel('Amount ($)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July'],
    'Income': [5000, 5200, 4800, 5300, 5400, 5500, 5600],
    'Expenses': [4000, 4200, 3900, 4100, 4300, 4400, 4500]
}
df = pd.DataFrame(data)

# Calculate Savings
df['Savings'] = df['Income'] - df['Expenses']

# Plotting the bar graph
plt.figure(figsize=(12, 6))
bar_width = 0.2
index = range(len(df))

# Bar positions
r1 = index
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Bar plots
plt.bar(r1, df['Income'], color='blue', width=bar_width, edgecolor='grey', label='Income')
plt.bar(r2, df['Expenses'], color='red', width=bar_width, edgecolor='grey', label='Expenses')
plt.bar(r3, df['Savings'], color='green', width=bar_width, edgecolor='grey', label='Savings')

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.title('Monthly Income, Expenses, and Savings')
plt.xticks([r + bar_width for r in range(len(df))], df['Month'])
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
